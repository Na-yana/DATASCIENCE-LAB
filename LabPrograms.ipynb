```python
#NAIVE BAYES ALGORITHM
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
X,y=load_iris(return_X_y=True)
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.2,random_state=0)
nb=GaussianNB()
y_pred=nb.fit(Xtrain,ytrain).predict(Xtest) 
print(y_pred)
x_new=[[5,5,4,4]]
y_new=nb.fit(Xtrain,ytrain).predict(x_new)
cm=confusion_matrix(ytest,y_pred)
print(cm)
print(y_new)
print(nb.score(Xtest,ytest))





```

```python
#SVM
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn import metrics
iris=load_iris()
x=iris.data
y=iris.target
Xtrain,Xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=1)
classifier=SVC(kernel="linear", random_state=0)
classifier.fit(Xtrain,ytrain)
y_pred=classifier.predict(Xtest)
print("Accuracy score",metrics.accuracy_score(ytest,y_pred))
sample=[[1,1,1,2]]
pred=classifier.predict(sample)
pred_v=[iris.target_names[p]for p in pred]
print(pred_v)




```

```python
#KMeans clustering
from sklearn. datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
iris=load_iris()
x=iris.data[:,:2]
y=iris.target
km = KMeans(n_clusters=3,init='k-means++',n_init=10,max_iter=300,tol=0.0001,verbose=0,random_state=21,copy_x=True,algorithm="auto")
km.fit(x)
centers=km.cluster_centers_
print(centers)
new_labels=km.labels_
print(new_labels)
print(y)
```

```python
#REGRESSION
from sklearn import datasets,linear_model
import matplotlib.pyplot as plt
import numpy as np
df=datasets.load_diabetes()
diabetes_x,diabetes_y=datasets.load_diabetes(return_X_y=True)
diabetes_x=diabetes_x[:,np.newaxis,2]
diabetes_x_train=diabetes_x[:-20]
diabetes_x_test=diabetes_x[-20:]
diabetes_y_train=diabetes_y[:-20]
diabetes_y_test=diabetes_y[-20:]
reg=linear_model.LinearRegression()
reg.fit(diabetes_x_train,diabetes_y_train)
diabetes_y_pred=reg.predict(diabetes_x_test)
from sklearn.metrics import mean_squared_error,r2_score
print("Coeeficients\n",reg.coef_)
print("Mean squared error:%.2f"%mean_squared_error(diabetes_y_test,diabetes_y_pred))
print("coefficient of determination:%.2f" %r2_score(diabetes_y_test,diabetes_y_pred))





 
```
